```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.scatter(x,y,c='b',marker='+')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter-Plot')
plt.show()
```

```python
plt.plot(x,y,'*-',color='blue')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.plot(x,y,'r*',color='cyan',linewidth=5,markersize='10')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.plot(x,y,c='r')
plt.title('Sub-Plot 1')
plt.subplot(2,2,2)
plt.plot(x,y,c='b')
plt.title('Sub-Plot 2')
plt.subplot(2,2,3)
plt.plot(x,y,c='g')
plt.title('Sub-Plot 3')
plt.subplot(2,2,4)
plt.plot(x,y,c='yellow')
plt.title('Sub-Plot 4')
plt.tight_layout()
plt.show()
```

```python
x = np.arange(1,11)
y = x**2
plt.plot(x,y,'-',color='pink',linewidth=5,markersize='10')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
x = np.arange(1,11,0.01)
y = np.sin(x)
plt.plot(x,y,'-',color='cyan',linewidth=2,markersize='2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine-wave')
plt.show()
```

```python
x = np.arange(1,11,0.01)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)
plt.figure(figsize=(10,6))
plt.subplot(3,1,1)
plt.plot(x,y_sin,'-',color='red',linewidth=2,markersize='2')
plt.title('Sine-wave')
plt.subplot(3,1,2)
plt.plot(x,y_cos,'-',color='blue',linewidth=2,markersize='2')
plt.title('Cosine-wave')
plt.subplot(3,1,3)
plt.plot(x,y_tan,'-',color='green',linewidth=2,markersize='2')
plt.title('Tangent-wave')
plt.tight_layout()
plt.show()
```

```python
x = np.array([2,5,4])
y = x
plt.bar(x,y,color='blue')
plt.title('Single Bar Plot')
plt.show()
```

```python
x = np.array([2,5,4])
y = x
x2 = np.array([3,7,9])
y2 = x
plt.bar(x,y,color='blue')
plt.bar(x2,y2,color='green')
plt.title('Double Bar Plot')
plt.show()
```

```python
x = np.array([2,5,4])
y = x
x2 = np.array([3,7,9])
y2 = x
x3 = np.array([11,24,1])
y3 = x/2
plt.bar(x,y,color='blue')
plt.bar(x2,y2,color='green')
plt.bar(x3,y3,color='red')
plt.title('Multiple Bar Plot')
plt.show()
```

```python
hist1 = np.random.randint(0,100,50)
plt.hist(hist1)
plt.title('Histogram')
plt.show()
```

```python
hist1 = np.random.randint(0,100,50)
plt.hist(hist1,bins=20)
plt.title('Histogram')
plt.show()
```

```python
box1 = [np.random.normal(0,std,100) for std in range(1,4)]
```

```python
plt.boxplot(box1,vert=True,patch_artist=True)
plt.title('Boxplot')
plt.show()
```

```python
plt.boxplot(box1,vert=True,patch_artist=False)
plt.show()
```

```python
labels = ['Python','Julia','C++','R']
size = [450,100,150,300]
colors = ['pink','lightblue','lightyellow','cyan']
explode = [0.15,0,0,0]

plt.pie(size,explode=explode,colors=colors,labels=labels,shadow=True,autopct='%1.1f%%')
plt.axis('equal')
plt.title('Pie-Chart')
plt.tight_layout()
plt.show()
```
