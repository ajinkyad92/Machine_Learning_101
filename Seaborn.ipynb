# Tips Dataset

```python
df_tips = sns.load_dataset('tips')
```

```python
df_tips.head()
```

```python
df_tips.info()
```

```python
df_tips['size']= df_tips['size'].astype('category')
```

```python
def dist_kde_plots(df):
    for i in df.columns:
        if df[str(i)].dtype == int or df[str(i)].dtype == float :
            sns.distplot(df[str(i)],kde =True)
            plt.show()
            
def dist_plots(df):
    for i in df.columns:
        if df[str(i)].dtype == int or df[str(i)].dtype == float :
            sns.distplot(df[str(i)],kde =False)
            plt.show()
```

```python
dist_plots(df_tips)
```

```python
dist_kde_plots(df_tips)
```

```python
sns.heatmap(df_tips.corr(),annot=True)
plt.show()
```
